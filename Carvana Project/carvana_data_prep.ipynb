# Dealing with lists in columns

The real issue here is: some of the columns have empty lists as they did not get normalized properly, these empty lists are NaNs but are not perceived as ones. Writing a function to get rid of lists:

```python
def list_opener(value):
    if value == []:
        # if empty list
        return np.NaN
    else:
        try:
            # if non-empty list
            return value[0]
        except:
            # if not a list
            return value


columns_with_lists = [
    "details.basics.poseData.installedPackages",
    "details.basics.poseData.installedOptions",
    "details.basics.poseData.installedOptionsData",
    "details.basics.poseData.standardEquipment"
]

for col in columns_with_lists:
    trimmed[col] = trimmed[col].apply(
        lambda x: list_opener(x))
```

# Checking for NaNs

```python
trimmed.isnull().sum().sort_values(ascending=False)
```

We have 329 rows, top 7 columns in the list above have too many NaNs. Lets drop them: 

```python
df = trimmed.drop(["details.basics.marketplaceDealerId",
                    "details.basics.poseData.installedPackages",
                    "details.basics.poseData.installedOptionsData",
                    "details.basics.poseData.installedOptions",
                    "details.basics.windowStickerUrl",
                    "details.basics.poseData.windowStickerUrl",
                    "details.basics.poseData.standardEquipment"], axis=1)

df.head()
```

```python
all(list(df.body_style == df.body_type))
```

...They do. And I wonder why. 

```python
df[df.body_style != df.body_type]
```

_SUV_ stands for _Sport_ _Utility_ _Vehicle_, therefore it's **technically** the same. I like short abbreivations more, so I am going to get rid of **"body_style"** column:

### To be converted:
- **seats** - to _int_
- **scraped_at** - to _datetime_

```python
df.seats = df.seats.astype(int)
df.scraped_at = df.scraped_at.astype(np.datetime64)
```

## Checking data for consistency:

```python
# I cut off the first two columns (url and id) since they are not as much 
# interesting for us:
for item in df.columns[2:]:                           
    print(item, sorted(list(df[item].unique())))
```

**horsepower**, **torque**, **displacement** - has "0" values. If a car drives, those cannot be _zeros_, so in reality they are _NaNs_. Let's change that:

```python
df.loc[df.horsepower == 0, 'horsepower'] = np.NaN
df.loc[df.torque == 0, 'torque'] = np.NaN
df.loc[df.displacement == 0, 'displacement'] = np.NaN
```

Additionally, some values in **make** need to be switched to a capitalized lower case for consistency:

```python
df.make = df.make.apply(lambda x: x.capitalize())
```
